```python
import pandas as pd
import logging

logging.basicConfig(level=logging.INFO)
```

```python
url = "https://id.wikipedia.org/wiki/Daftar_negara_penghasil_komoditas_agrikultural_terbesar"
```

```python
def scrape(url):
  logging.info(f"Scraping website url: '{url}'....")
  return pd.read_html(url, header=None)
```

```python
df = scrape(url)[0]
```

```python
df
```

```python
df.to_excel ('Komoditas_Dunia.xlsx')
```
